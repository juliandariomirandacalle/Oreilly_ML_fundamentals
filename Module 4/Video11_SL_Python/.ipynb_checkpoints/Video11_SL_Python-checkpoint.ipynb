# Modeling Denial of Service Attacks

##  Introduction to the Case Study

**Business Context.** You are the Machine Learning specialist of the United States Federal Statistical System (FSS) and you have been hired to find patterns in computer attacks that have been happening and registered in a tabular data set.

**Business Problem.** You want to be able to distinguish Denial of Service (DoS) computer attacks from generic ones that may affect the availability of Internet services of the organization. 

**Analytical Context.** The data to be processed has been obtained by aggregating data provided by the company's technology department, which includes network flow data produced by nine different types of attacks within everyday user activities. Those attacks were grouped into two main categories, Denial of Services (DoS) and generic attacks. Using your Machine Learning skills, build a binary classifier that models the Denial of Services to be able to distinguish these attacks from other less relevant attack behaviors.

## Importing the packages for data analysis

We will use six packages: `pandas`, `matplotlib`, `seaborn`, `datetime`, `sklearn` and `statsmodels`. Let's import these packages using the `import` keyword. We will rename `pandas` to `pd`, `matplotlib` to `plt` and `seaborn` to `sns`, using the `as` keyword:

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import statsmodels.api as sm
````

We are going to read a .csv file that contains information about different cybersecurity events that have occurred in a given time frame. The data set has the following attributes:

1. **Start time:** Recorded start date of the attack in *timestamp* format.
2. **Last time:** Recorded end date of the attack in *timestamp* format.
3. **Attack category:** type of attack registered.
4. **Protocol:** protocol used for the attack.
5. **Source IP:** registered IPv4 (Internet protocol) address where the attack came from.
6. **Source Port:** registered logical port where the attack came from.
7. **Destination IP:** Registered destination IPv4 (Internet protocol) address of the attack.
8. **Destination Port:** Registered logical destination port of the attack.
9. **Destination Port Service:** service provided by the destination port.
10. **Attack Reference:** reference in the Common Vulnerabilities and Exposures (CVE) of the registered attack type.

The following attack types are included in the data set:

- **GENERIC:** groups computer attacks for reconnaissance, exploitation and backdoors, among others.
- **DOS (DoS - Denial of Services):** is a computer attack that causes interruptions in resources and services. This attack aims to overload memory to keep services extremely busy, preventing requests from authorized users from accessing a device and/or unauthorized ones from accessing services.

To get a better idea of the data available, let's first look at the data from the US Federal Statistical System (FSS), found in the `cybersecurity_attacks.csv` file within the `data_files` folder:

````python
df_interest = pd.read_csv('data_files/cybersecurity_attacks.csv')  # Import the atacks data set
df_interest.head(3)                                                # Show the first three rows
````

Let's see the distribution of computer attacks in the data:

````python
# Plotting the distribution of computer attacks
plt.figure(figsize=(18,6))                                            # Creating the  plotting object with a fixed size
sns.barplot(x=df_interest['Attack category'].value_counts().index,    # Plotting the number of samples per computer attach
            y=df_interest['Attack category'].value_counts())
plt.xlabel('Attack Category')                                         # Defining the x-axis label
plt.ylabel('Count')                                                   # Defining the y-axis label
plt.title('Number of attacks per Attack caterogy')                    # Defining the plot title
plt.grid(True)                                                        # Showing the grid
plt.show()                                                            # Showing the plotting object
````

As we can see, DoS and Generic attacks are relatively evenly distributed in the data.

Now let's transform the dates into the corresponding format and calculate the duration, in seconds, of the recorded computer attacks:

````python
df_interest['Start time'] = pd.to_datetime(df_interest['Start time'])       # Formatting the start date into datetime
df_interest['Last time'] = pd.to_datetime(df_interest['Last time'])         # Formatting the start date into datetime
duration_attacks = (df_interest['Last time'] - df_interest['Start time'])   # Subracting the start date from the end date
df_interest['Duration'] = (duration_attacks.dt.seconds).astype(int)         # Calculating the attacks' duration in seconds
````

Now let's take a closer look at the dispersion of the destination (target) port of the attacks with emphasis on the first 200 well-known ports and that happened on the morning of February 15th, 2018:

````python
# Defining some plotting parameters
time_format = '%y-%m-%d %H:%M:%S'                                    # Datetime format to be used
data = df_interest[(df_interest['Destination Port']<=200)]           # Keep attacks pointing out to ports under the 200

# Scatter plot of the destination port for DoS and generic attacks
plt.figure(figsize=(18,7))                                           # Creating the  plotting object with a fixed size
sns.scatterplot(x='Start time',                                      # Plotting the destination plot w.r.t the computer attack
                y='Destination Port', 
                hue='Attack category',
                data=data)
plt.xlim(left=datetime.strptime('15-02-18 00:00:00', time_format),   # Framing the plot within the defined time interval
         right=datetime.strptime('15-02-18 13:00:00', time_format))
plt.grid(True)                                                       # Showing the grid
plt.show()                                                           # Showing the plotting object
````

We can notice that there is a higher density of computer attacks on specific destination ports. This is because some of those ports below 1024 present common services, such as:

| Port | Service | Service definition                                                                                                       |   |   |
|------|---------|--------------------------------------------------------------------------------------------------------------------------|---|---|
| 22   | SSH     | The Secure Shell Protocol (SSH) is a cryptographic protocol that enables secure communcations through unsecure networks. |   |   |
| 53   | DNS     | Domain Name System (DNS) is a hierarchical and distributed naming system for devices and services in Internet networks.  |   |   |
| 80   | HTTP    | Hypertext Transfer Protocol (HTTP) is a foundational protocol of hypermedia information systems for data communication.  |   |   |
| 143  | IMAP    | Internet Message Access Protocol (IMAP) is a protocol used for email messages retrieval from mail servers                |   |   |

To see this relationship between attacks and source and destination ports in more detail, we can make a strip plot like this:

````python
# Plotting the source ports w.r.t. the computer attacks
plt.figure(figsize=(18,7))            # Creating the  plotting object with a fixed size
sns.stripplot(x='Attack category',    # Plotting a strip plot of the source ports
              y='Source Port',
              data=df_interest)
plt.show()                            # Showing the plotting object
````

````python
# Plotting the destination ports w.r.t. the computer attacks
plt.figure(figsize=(18,7))            # Creating the  plotting object with a fixed size
sns.stripplot(x='Attack category',    # Plotting a strip plot of the destination ports
              y='Destination Port',
              data=df_interest)
plt.show()                            # Showing the plotting object
````

Let's look at the pair of graphs above in detail. The attacks we are analyzing seem evenly distributed across all source ports, from 0 to 65,535. There is no clear pattern beyond the fact that there is a slightly greater concentration on the first 1,000 source ports.

On the other hand, the story with the destination ports is very different. As we can notice, the highest concentration of attacks occurs on ports below 10,000, possibly the best-known ports (ports below 1,024). Some Denial of Service attacks can be identified on specific ports above 10,000, perhaps because (a) the attacker has identified that some Internet services were relocated or (b) because the attacker ran random tests on some of these ports. We can see that it is a little more common to have DoS attacks on higher ports than it is for generic attacks, which have a large concentration on lower ports.

If we used the destination port to distinguish the types of computer attacks presented, we would have a difficult task determining a simple linear boundary. Let's delve into the development of a binary classifier using Logistic Regression to determine this separation in a more elaborated way.

## Logistic Regression

We are going to build a classification model that allows us to identify which cybersecurity events correspond to Denial of Services (DoS) and which correspond to generic attacks. We will do this by building a Logistic Regression model.

The Logistic Regression model is a modified version of the standard linear regression model that deals with classification problems, where the outcomes are binary variables (for binary classification problems). To understand the concept of logistic regression, let's first introduce the idea of probabilities. The probability of an event is the relationship between its probability of success and its probability of failure. In our case, we are going to model the *odds* of a cybersecurity event being a Denial of Services:
$$
\text{Odds}(\text{DoS}) = \frac{p}{1-p}. \tag{Equation 1}
$$

Where $p$ is the probability of DoS attack.

It turns out that *odds* are easier to work with than probabilities. To solve classification problem let's first assume that the probabilities are a function of the duration of the attacks:

\begin{equation}
\log(\text{Odds}) = \beta_0 + \beta_1*\text{duration}. \tag{Equation 2}
\end{equation}

Let's train our first Logistic Regression (LR) model considering the attack duration.

### LR model 1 - Features: attack duration

Let's first structure our data set for classification. We are going to extract all the numerical components of the data set and define the label for the classification as class 1, for attacks that correspond to Denial of Services and class 0 for those that correspond to generic attacks:

````python
#  Defining the classification labels as:
## Class 1:Denial of Services attacks
## Class 0: generic (other) attacks
df_interest_cl = df_interest.copy()
df_interest_cl['label'] = 0
df_interest_cl.loc[df_interest_cl['Attack category'] == 'DOS', 'label'] = 1

# Getting only the numerical features
df_interest_cl = df_interest_cl._get_numeric_data()
df_interest_cl.head(3)
````

````python
# Building the data set for our first LR model:
## Features: the duration of the attacks
## Label: two classes - Class 0 for generic attacks and Class 1 for all DoS attacks
df_log1 = pd.DataFrame(columns=['Duration','label'])
df_log1['label'] = df_interest_cl['label']
df_log1['Duration'] = df_interest_cl['Duration']
df_log1['Intercept'] = 1
df_log1.head(3)
````

We have added an intercept column that has a single value. This column will allow us to vary the intercept as the LR model is trained. Now we are going to split the data into training and testing. This will allow us to measure the performance of the model in a more objective way. We are going to take 80% of the observations for training and 20% for testing:

````python
# Split the data set into two subsets: training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(df_log1.drop('label', axis=1),  # Features
                                                    df_log1['label'],               # Label
                                                    test_size = 0.2,                # Testing percentage (20%)
                                                    random_state = 0)               # Random state for replication purposes
````

Let's train the LR model using the `statmodels`' API:

````python
# Defining and training the LR model
logit = sm.Logit(y_train, X_train)    # Defining the LR model object with the given features and target variables
logit_res_1 = logit.fit()             # Fitting the LR model
print(logit_res_1.summary())          # Printing the model's summary
````

This indicates that when the Duration increases by one (1) second, all other things being equal, the probability of it being a denial of service attack increases by 3.32% (0.0332 coefficient for the attack duration)!

To measure the classification performance with the test observations, let's plot the confusion matrix with the `confusion_matrix()` method:

|    |    |
|----|----|
| TN | FP |
| FN | TP |

Where a True Positive (TP) corresponds to a DoS attack that was classified as such by the model, and a True Negative (TN) corresponds to a generic attack classified as such by the model.

````python
# Structuring the confusion matrix (for binary classification)
classification_threshold = 0.5                                      # Classification threshold for binarization
y_pred = logit_res_1.predict(X_test) >= classification_threshold    # Binarizing the probabilities with the given threshold
y_pred = (y_pred).astype(int).values                                # Extracting the binarized values
print(confusion_matrix(y_test, y_pred))                             # Printing the confusion matrix
````

We can see that the logistic regression model performs well for Class 1 (treu positives are 4887). However, all generic attacks (Class 0) were classified as DoS attacks (4002 events). Hence, the LR model trained with the duration as the main feature classifies every attack as a Denial of Services.

In the following section, we will include the source port `Source Port` as a second feature.

### LR model 2 - Features: attack duration and source port

````python
# Extracting the duration and source port features and defining the intercept
df_log2 = df_interest_cl[['Duration','Source Port','label']].copy()
df_log2['Intercept'] = 1

# Split the data set into two subsets: training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(df_log2.drop('label', axis=1),  # Features
                                                    df_log2['label'],               # Label
                                                    test_size = 0.2,                # Testing percentage (20%)
                                                    random_state = 0)               # Random state for replication purposes

# Defining and training the LR model
logit = sm.Logit(y_train, X_train)    # Defining the LR model object with the given features and target variables
logit_res_2 = logit.fit()             # Fitting the LR model
print(logit_res_2.summary())          # Printing the model's summary
````

Let's calculate the confusion matrix of the updated classifier:

````python
# Structuring the confusion matrix (for binary classification)
classification_threshold = 0.5                                      # Classification threshold for binarization
y_pred = logit_res_2.predict(X_test) >= classification_threshold    # Binarizing the probabilities with the given threshold
y_pred = (y_pred).astype(int).values                                # Extracting the binarized values
print(confusion_matrix(y_test, y_pred))                             # Printing the confusion matrix
````

In the confusion matrix we note the following:
- TP: 4,343 DoS attacks were classified as such by the model
- TN: 1,907 generic attacks were classified as such by the model
- FP: 2,095 generic attacks were classified as DoS by the model
- FN: 544 DoS attacks were classified as generic attacks by the model

The classifier has improved its performance in classifying generic attacks and no longer classifies all samples as coming from the positive class.

Now let's include the destination port, fit a Logistic Regression model and check its performance again (see the following section).

### LR model 3 - Features: attack duration, source and destination ports

````python
# Extracting the duration and source port features and defining the intercept
df_log3 = df_interest_cl.copy()
df_log3['Intercept'] = 1

# Split the data set into two subsets: training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(df_log3.drop('label', axis=1),  # Features
                                                    df_log3['label'],               # Label
                                                    test_size = 0.2,                # Testing percentage (20%)
                                                    random_state = 0)               # Random state for replication purposes

# Defining and training the LR model
logit = sm.Logit(y_train, X_train)    # Defining the LR model object with the given features and target variables
logit_res_3 = logit.fit()             # Fitting the LR model
print(logit_res_3.summary())          # Printing the model's summary
````

````python
# Structuring the confusion matrix (for binary classification)
classification_threshold = 0.5                                      # Classification threshold for binarization
y_pred = logit_res_3.predict(X_test) >= classification_threshold    # Binarizing the probabilities with the given threshold
y_pred = (y_pred).astype(int).values                                # Extracting the binarized values
print(confusion_matrix(y_test, y_pred))                             # Printing the confusion matrix
````

We managed to reduce false negatives by 4 and increase true positives by the same value. This represents a slight improvement in the model's performance.

The overall performance of the model w.r.t common binary classification metrics can be calculated using the `classification_report()` function:

````python
# Printing the general classification report for LR
print(classification_report(y_test, y_pred))
````

The performance of the model measured with the F1-score (macro average), for instance, is around 70%, which is an acceptable performance given the fact that classes are almost balanced.

Once we have fitted the model and tested it with our test dataset, we are ready to use it. Let's review, for example, the likelihood of an attack being DoS given that the origin port is 1043, the destination port 53 (DNS service) and the attack lasted 5 seconds:

````python
# Out-of-sample observation definition with the given features
sample = {
    'Source Port': 1043,
    'Destination Port': 53,
    'Duration': 5,
    'Intercept': 1,
}

# Printing the likelihood of an attack being DoS given the features above
probas = logit_res_3.predict(list(sample.values()))[0]*100      # Estimating the likelihood given the features
print('DoS attack probability:', str(round(probas, 2)) + '%')   # Printing the results
````

The likelihood of an attack being DoS coming from the port 1042 and going to port 53 with a duration of 5 seconds is 70.11%.
